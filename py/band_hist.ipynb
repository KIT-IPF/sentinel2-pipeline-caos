```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
%config IPCompleter.greedy=True

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
import os
import sys

sns.set_style('whitegrid')
```

```python
df = pd.read_csv('./processed_data/merged_dataset.csv', sep=';', index_col=0, na_values='NA')

bands = ['b1', 'b2', 'b3', 'b4', 'b5', 'b6', 'b7', 'b8', 'b9', 'b10', 'b11', 'b12', 'b8a']
```

```python
for b in bands:
    print(b)
    sns.distplot(df[df['b10']<600][b].dropna(), kde=False, rug=True, norm_hist=True)
    plt.savefig('./plots/'+b+'_hist.pdf', dpi=300)
    plt.show()
```

```python
for b in bands[:6]:
    print(df[df[b]>4000][['Date','Station']])
```

```python
ds = pd.read_csv('stations-dates_to_ignore.txt', delimiter='\t')

df = df[~(df['Date'].isin(ds['Date']) & df['Station'].isin(ds['Station']))]
    

  
```

```python
for b in bands:
    print(b)
    sns.distplot(df[df['b10']<600][b].dropna(), kde=False, rug=True, norm_hist=True)
    plt.savefig('./plots/'+b+'_hist_filtered.pdf', dpi=300)
    plt.show()
```

```python
df.to_csv('./processed_data/merged_dataset.csv', sep=';', na_rep='NA')
```
